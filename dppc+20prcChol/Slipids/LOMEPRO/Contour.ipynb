```python
import numpy as np
import matplotlib.pyplot as plt
data01 = np.loadtxt('thickness.out_avg.dat')
fig, ax=plt.subplots(figsize=(10, 8))
cp = ax.contourf(data01, cmap='bwr')
fig.colorbar(cp, label='Thickness (nm)') # Add a colorbar to a plot
ax.set_title('1 ns')
plt.show()
fig.savefig('Thickness.jpg', dpi=200)
```
